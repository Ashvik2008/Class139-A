```
import pandas as pd 
import numpy as np 

df1=pd.read_csv('tmdb_5000_credits.csv')
df2=pd.read_csv('tmdb_5000_movies.csv')
```

```
df1.head()
```

```
df2.head()
```

```
df1.columns = ['id','tittle','cast','crew']
df2= df2.merge(df1,on='id')

df2.head(5)
```

# Demographic Filtering

```
C = df2['vote_average'].mean()
print(C)
```

```
m = df2['vote_count'].quantile(0.9)
print(m)
```

```
print(df2.shape)
q_movies = df2.copy().loc[df2['vote_count'] >= m]
print(q_movies.shape)
```

```
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)

q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
q_movies['score']
```

V = number of votes for the movie.
m = Min number of vote required to be listed in chart.
R = average rating of individual movie.
C = Mean vote across the whole report(same for all).

```
q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```
import plotly.express as px

fig = px.bar((q_movies.head(10).sort_values('score', ascending=True)), x="title", y="score", orientation='v')
fig.show()
```
